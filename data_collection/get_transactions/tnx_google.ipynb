```python
def find_avg_time_diff(time_col):
    if len(time_col) == 0:
        return 0
    diff = 0
    for i in range(1, len(time_col)):
        diff += abs(float(time_col[i]) - float(time_col[i - 1]))

    return diff/(len(time_col)*60)

def set_received_properties(df):
    erc_total_ether_rec.append((df['value'].sum()))
    erc_min_val_rec.append(df['value'].min())
    erc_max_val_rec.append((df['value'].max()))
    erc_avg_val_rec.append((df['value'].mean()))
    erc_avg_time_between_rec.append(find_avg_time_diff(df['timeStamp']))
    received_tnx.append(len(df))
    erc_uniq_rec_addr.append(df['from'].nunique())
    
def set_sent_properties(df):
    erc_total_ether_sent.append(df['value'].sum())
    erc_min_val_sent.append(df['value'].min())
    erc_max_val_sent.append(df['value'].max())
    erc_avg_val_sent.append(df['value'].mean())
    sent_tnx.append(len(df))
    erc_uniq_sent_addr.append(df['to'].nunique())
    erc_avg_time_between_sent.append(find_avg_time_diff(df['timeStamp']))
    contractAddress_tnx = df.where(df['contractAddress'] != "")
    if contractAddress_tnx.empty:
        min_val_sent_to_contract.append(0)
        max_val_sent_to_contract.append(0)
        avg_val_sent_to_contract.append(0)
        total_ether_sent_contracts.append(0)
    else:
        min_val_sent_to_contract.append(contractAddress_tnx['value'].min())
        max_val_sent_to_contract.append(contractAddress_tnx['value'].max())
        avg_val_sent_to_contract.append(contractAddress_tnx['value'].mean())
        total_ether_sent_contracts.append(contractAddress_tnx['value'].sum())
```

```python
def fill_with_zeros_sent():
    erc_total_ether_sent.append(0)
    erc_min_val_sent.append(0)
    erc_max_val_sent.append(0)
    erc_avg_val_sent.append(0)
    erc_uniq_sent_addr.append(0)  
    sent_tnx.append(0)
    erc_avg_time_between_sent.append(0)
    min_val_sent_to_contract.append(0)
    max_val_sent_to_contract.append(0)
    avg_val_sent_to_contract.append(0)
    total_ether_sent_contracts.append(0)

def fill_with_zeros_received():
    erc_total_ether_rec.append(0)
    erc_min_val_rec.append(0)
    erc_max_val_rec.append(0)
    erc_avg_val_rec.append(0)
    erc_uniq_rec_addr.append(0)
    received_tnx.append(0)
    erc_avg_time_between_rec.append(0)

def fill_with_nan():
    erc_total_ether_rec.append(np.nan)
    erc_total_ether_sent.append(np.nan)
    erc_uniq_sent_addr.append(np.nan)
    erc_uniq_rec_addr.append(np.nan)
    erc_avg_time_between_sent.append(np.nan)
    erc_avg_time_between_rec.append(np.nan)
    erc_min_val_rec.append(np.nan)
    erc_max_val_rec.append(np.nan)
    erc_avg_val_rec.append(np.nan)
    erc_min_val_sent.append(np.nan)
    erc_max_val_sent.append(np.nan)
    erc_avg_val_sent.append(np.nan)
    total_ether_balance.append(np.nan)
    time_diff_between_first_and_last.append(np.nan)
    number_of_created_contracts.append(np.nan)
    min_val_sent_to_contract.append(np.nan)
    max_val_sent_to_contract.append(np.nan)
    avg_val_sent_to_contract.append(np.nan)
    total_transactions.append(np.nan)
    total_ether_sent_contracts.append(np.nan)
    received_tnx.append(np.nan)
    sent_tnx.append(np.nan)
```
